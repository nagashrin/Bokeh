```python
# Bokeh libraries
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import reset_output
from bokeh.palettes import Viridis5
from bokeh.models import HoverTool
import pandas as pd
import math
```

```python
clo_by_reason = pd.read_excel('S:\\TANF Reports\\!2019 TANF Reports\\2-February 2019\Data_viz\\Data Visualization - February 2019.xlsx', 
                              sheet_name='closure by Reason_data')

output_notebook()
# create a new dataframe with the columns required for visualization
df2 = pd.DataFrame({'count': clo_by_reason.groupby(['Reason']).size()}).reset_index()


reason = df2['Reason'].tolist()
count = df2['count'].tolist()

# Bokeh's mapping of column names and data lists
source = ColumnDataSource(data=dict(reason=reason, count=count, color= Viridis5))

# Bokeh's convenience function for creating a Figure object
p = figure(x_range=reason, y_range=(0, 65), plot_height=350, title="Counts",
           toolbar_location='right')

# Render and show the vbar plot
p.vbar(x='reason', top='count', width=0.9, color='color', source=source)

# Add a hover tool referring to the formatted columns
hover = HoverTool(tooltips = [('Count', '@count')])

# Add the hover tool to the graph
p.add_tools(hover)

# Show the plot
show(p)

```

```python
clo_by_reason = pd.read_excel('S:\\TANF Reports\\!2019 TANF Reports\\2-February 2019\Data_viz\\Data Visualization - February 2019.xlsx', 
                              sheet_name='closure by Reason_data')

output_notebook()
# create a new dataframe with the columns required for visualization
df_2 = pd.DataFrame({'count': clo_by_reason.groupby(['local office','Reason']).size()}).reset_index()
df_2 = df_2.pivot_table(index = ['local office'], columns= 'Reason', values = 'count').reset_index()

reasons = ['Earnings Exceeds FPL',
       'Failed to meet Initial Eligibility', 'Failed to meet participation',
       'Failed to provide info', 'Failed to verify income',
       'Minor Parent Turned 18 yr', 'No Child in Home', 'Not State Resident',
       'Other-Needs Related', 'Other-Non Need', 'Progressive Sanction',
       'Requested closure', 'System -SSI Benefits', 'TEA Close Reached 24 Mo',
       'TEA EP Non-compliance', 'TEA Excess Income',
       'TEA Extra Payment-Earning', 'TEA Extra Payments-Request',
       'TEA Lifetime limit', 'TM form not returned', 'Unable to locate',
       'Unable to locate-mail', 'Work Pays time limit met']
counts = sum(zip(df_2['Earnings Exceeds FPL'],
                 df_2['Failed to meet Initial Eligibility'],df_2['Failed to meet participation'],
                 df_2['Failed to provide info'],df_2['Failed to verify income'],
                 df_2['Minor Parent Turned 18 yr'],df_2['No Child in Home'],
                 df_2['Not State Resident'],df_2['Other-Needs Related'],
                 df_2['Other-Non Need'],df_2['Progressive Sanction'],
                 df_2['Requested closure'],df_2['System -SSI Benefits'],
                 df_2['TEA Close Reached 24 Mo'],df_2['TEA EP Non-compliance'],
                 df_2['TEA Excess Income'],df_2['TEA Extra Payment-Earning'],
                 df_2['TEA Extra Payments-Request'],df_2['TEA Lifetime limit'],
                 df_2['TM form not returned'],df_2['Unable to locate'],
                 df_2['Unable to locate-mail'],df_2['Work Pays time limit met']), ())
localoffices = df_2['local office'].tolist()

# Bokeh's mapping of column names and data lists
source = ColumnDataSource(data=dict(localoffices=localoffices, reasons=reasons, counts=counts, color= Viridis5))

x = [(localoffice, reason) for localoffice in localoffices for reason in reasons]

# Bokeh's convenience function for creating a Figure object
p = figure(x_range=FactorRange(*x), y_range=(0, 5), plot_height=350, title="Counts",
           toolbar_location='right')

# Render and show the vbar plot
p.vbar(x='x', top='counts', width=0.9, color='color', source=source)

# Add a hover tool referring to the formatted columns
hover = HoverTool(tooltips = [('Count', '@counts')])

# Add the hover tool to the graph
p.add_tools(hover)

# Show the plot
show(p)
```
